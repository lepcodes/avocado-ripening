```python
import pandas as pd
import shutil
import cv2
import os
```

```python
dataset = pd.read_excel("C:/Users/Luis/Documents/ML-AI-Projects/avocado-ripening/data/external/Hass Avocado Ripening Photographic Dataset/Avocado Ripening Dataset.xlsx")
```

```python
num_samples = 250
dataset_batch = dataset.loc[dataset['Sample'] <= num_samples]
```

```python
image_path = "C:/Users/Luis/Documents/ML-AI-Projects/avocado-ripening/data/external/Hass Avocado Ripening Photographic Dataset/Avocado Ripening Dataset"
dump_csv_path = "C:/Users/Luis/Documents/ML-AI-Projects/avocado-ripening/data/raw/dataset_batch_1"
dump_image_path = "C:/Users/Luis/Documents/ML-AI-Projects/avocado-ripening/data/raw/dataset_batch_1/images"

os.makedirs(dump_image_path, exist_ok=True)
for image in dataset_batch["File Name"]:
    if os.path.exists(image_path + "/" + image + ".jpg"):
        if not os.path.exists(dump_image_path + "/" + image + ".jpg"):
            shutil.copy(image_path + "/" + image + ".jpg", dump_image_path + "/" + image + ".jpg")
        else:
            print("Image already exists")
    else:
        print("File not found")

if not os.path.exists(dump_csv_path):
    os.makedirs(dump_csv_path)

dataset_batch.to_csv(dump_csv_path + "/dataset_batch_1.csv", index=False)
```
